```python
class person:
    def __init__(self, name , age):
        self.name=name
        self.age=age
    def get_info(self):
        print(f"person name is {self.name}, person age is {self.age}")

person_1=person("Ahmed",21)

person_1.get_info()


```

```python
class dog:
    def __init__(self,name):
        self.name=name
    def bark(self):
        print(f"{self.name} is barking")

dog_1=dog("Max")
dog_2=dog("Roy")

dog_1.bark()
dog_2.bark()
```

```python
class person:
    def __init__(self, name,age):
        self.name=name
        self.age=age
    def speaks(self):
        print(f"{self.name} is speaking")

class doctor(person):
    def __init__(self, name, age):
        super().__init__(name, age)
    def doctor_speaks(self):
        print(f"Dr, {self.name} is speaking")

doctor_1= doctor("Ahmed",21,)
doctor_1.doctor_speaks()
doctor_1.speaks()
    
```

```python
class student:
    def __init__(self,name,grade):
        self.name=name
        self.grade=grade
    def update_grade(self,new_grade):
        self.grade=new_grade
    def show_info(self):
        print(f"student name is {self.name},grade is {self.grade}")
    

student_1=student("Ahmed",90)
student_1.show_info()
```

```python
class flyer:
    def __init__(self, wing_size):
        self.wing_size = wing_size
    def fly(self):
        return "can fly"

class swimmer:
    def __init__(self, fin_size):
        self.fin_size = fin_size
    def swim(self):
        return "can swim"

class duck(flyer, swimmer):
    def __init__(self, wing_size, fin_size):
        flyer.__init__(self, wing_size)  
        swimmer.__init__(self, fin_size)
    def get_info(self):  
        print(f"wing size: {self.wing_size} fin size: {self.fin_size}")

duck_obj = duck(10, 5)  
print(duck_obj.fly())  
print(duck_obj.swim())  
print(duck_obj.swim(), "and", duck_obj.fly())  
print(duck_obj.get_info())  


```
