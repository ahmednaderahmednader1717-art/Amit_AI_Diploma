```python
def get_numbers_type(numbers):
    even_numbers = []
    odd_numbers = []
    
    for i in numbers:
        if i % 2 == 0:
            even_numbers.append(i)
        else:
            odd_numbers.append(i)
    return even_numbers, odd_numbers

def get_even_sum(even_numbers):
    sum = 0
    for i in even_numbers:
        sum+= i
    return sum

def count_even(even_numbers):
    counter =0
    for i in even_numbers:
        counter= counter +1
    return counter
    

# ---------------- main ----------------
x = int(input("please enter the list length: "))
list_1 = []

for i in range(x):
    num = int(input(f"Enter number {i+1}: "))
    list_1.append(num)

even_numbers, odd_numbers = get_numbers_type(list_1)

print("even_numbers:", even_numbers)
print("odd_numbers:", odd_numbers)

even_sum = get_even_sum(even_numbers)
even_count = count_even(even_numbers)

print("sum of even numbers =", even_sum)
print("count of even numbers =", even_count)

if even_count > 0:
    avg = even_sum / even_count
    print("average of even numbers =", avg)
else:
    print("no even numbers to average")
```

```python
grade = int(input("Enter your grade: " ))
if grade < 0 or grade > 100:
    print("Invalid Grade")
else:
    if grade >= 90:
        print("Your Grade is 'A'")
    elif grade >= 80:
        print("Your Grade is 'B'")
    elif grade >= 70:
        print("Your Grade is 'C'")
    elif grade >= 60:
        print("Your Grade is 'D'")
    else:
        print("You have failed.")
```
