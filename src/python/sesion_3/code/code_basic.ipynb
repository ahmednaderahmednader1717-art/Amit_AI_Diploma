```python
def addtion(x,y):
    return x + y


x= int(input("please enter the first number: "))
y = int(input("please enter the second number: "))
z= addtion(x,y)
print(z)
```

```python
def sutraction(x,y):
    return x-y

x= int(input("please enter the first number: "))
y = int(input("please enter the second number: "))
z= sutraction(x,y)
print(z)
```

```python
def multiplication(x,y):
    return x*y

x= int(input("please enter the first number: "))
y = int(input("please enter the second number: "))
z= multiplication(x,y)
print(z)
```

```python
def division(x,y):
    return x/y

x= int(input("please enter the first number: "))
y = int(input("please enter the second number: "))
z= division(x,y)
print(z)
```

```python
operations = input("Enter the operations")

x= int(input("please enter the first number: "))
y = int(input("please enter the second number: "))

if operations == "+":
    result = addtion(x,y)
elif operations == "-":
    result = subtraction(x,y)
elif operations == "*":
    result = multiplication(x,y)
elif operations == "/":
    result = division(x,y)
else:
    print("Invalid operation")

print(result)

```
