```python
#task one

def process_email(email):

    if email.count("@") != 1:
        return "Invalid email"

    local_part, rest = email.split("@")

    if "." not in rest:
        return "Invalid email"

    username = local_part

    domain = rest.rsplit(".", 1)[0]

    if email.endswith(".com"):
        domain_type = "Commercial Domain"
    elif email.endswith(".edu"):
        domain_type = "Educational Domain"
    else:
        domain_type = "Other Domain"

    print(f" Username: {username},\n Domain: {domain},\n Domain Type: {domain_type}") 



email = input("Please Enter Your Email: ")
result = process_email(email)
print(result)
```
