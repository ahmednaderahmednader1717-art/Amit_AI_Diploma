```python
class student:
    def __init__(self,first,grade,age):
        self.first = first
        self.grade = grade
        self.email = first+'@company.com'
        self.age = age

    def get_info(self):
        return f"Name: {self.first}, Grade: {self.grade}, Email: {self.email}, Age: {self.age}"

std_1 = student("Ahmed","A+",21)

print(student.get_info(std_1))

```

```python
class employee:
    num_of_employees=0
    raise_amount=1.05
    def __init__(self,first,last,pay):
        self.first=first
        self.last=last
        self.pay=pay
        self.email=first+last+"@gmail.com"
        employee.num_of_employees += 1
    def fullname(self):
        return f"{self.first}, {self.last}"
    def apply_raise(self):
        self.pay=int(self.pay*self.raise_amount)
    def get_info(self):
        print(f"\nName: {self.first} {self.last}\nSalary: {self.pay}\nEmail: {self.email}")

print(employee.num_of_employees)

manger=employee("Ahmed","Nader",20000)

emp_1=employee("Kareem","Mohamed",3000)

print(employee.num_of_employees)

manger.raise_amount=3.0

emp_1.raise_amount=2

print("before raise:")

employee.get_info(manger)
employee.get_info(emp_1)

employee.apply_raise(manger)
employee.apply_raise(emp_1)

print("\nafter raise:")
employee.get_info(manger)
employee.get_info(emp_1)

```

```python
class animal:       #single inheritance
    def speak(self):
        return "Animal speaks"

class dog:
    def bark(self):
        return "Dog barks"


a=animal()
animal.speak(a)

d=dog()
dog.bark(d)
```

```python
class flyer: #multiple inheritence
    def fly(self):
        return "can fly"
class swimmer:
    def swim(self):
        return "can swim"
class duck(flyer,swimmer):
    pass
duck = duck()
print(duck.fly())
print(duck.swim())
```

```python
class grandparent:      # multilevel inheritance 
    def say_hello(self):
        return "Hello from Gedo"
class parent(grandparent):
    def say_hi(self):
        return "Hi from baba"
class child(parent):
    def say_hey(self):
        return "hey from son"
    
g=grandparent()
p=parent()
c=child()

print(child.say_hey(c))
print(parent.say_hi(p))
```

```python
class vehicle:     #hierarchical
    def start(self):
        return "engine started"
class car(vehicle):     
    def drive(self):
        return "car is driving"
class bike(vehicle):     
    def ride(self):
        return "bike is driving"

```

```python
class vehicle:
    def __init__(self,name,model):
        self.name=name
        self.model=model
    def vehicle_info(self):
        print(f"the vihicle name: {self.name}, the model: {self.model}\n")

class car(vehicle):
    def __init__(self, name, model,doors_number):
        super().__init__(name, model)
        self.doors_number=doors_number
    def cars_info(self):
        print(f"the model: {self.model}, doors number: {self.doors_number}")

car_1=car("BMW","320i",4)
vehicle_1=vehicle("BMW","320i")

vehicle.vehicle_info(vehicle_1)
car.cars_info(car_1)

```
