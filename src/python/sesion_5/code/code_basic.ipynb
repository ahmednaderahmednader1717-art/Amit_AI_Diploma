```python
def calculator(x,y,operations):
    
    if operations == "+":
        return x+y
    elif operations == "-":
        return x-y
    elif operations == "*":
        return x*y
    elif operations == "/":
            try:
                return x/y 
            except Exception:
                print("erorr: you cant devide by zero")
    else:
        print("Invalid operator")

operations = input("Enter the operations")
x= int(input("please enter the first number: "))
y = int(input("please enter the second number: "))
calculator(x,y,operations)


```
